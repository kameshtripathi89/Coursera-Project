```python
df=pd.read_excel("Online Retail.xlsx")
```

```python
df.head(6)
```

```python
df.replace("?", np.nan)
```

```python

df.isnull().sum()
```

```python

df["CustomerID"].value_counts()
```

```python

df.dropna(subset=["Quantity","UnitPrice"])
```

```python
df.shape
```

```python
df.drop(df[(df["Quantity"] < 0) | (df["UnitPrice"] < 0)].index, inplace=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Date'] = df['InvoiceDate'].dt.date
```

```python

df.head(5)
```

```python
df_group = df.groupby('Date')['Quantity'].sum()   # sums the total quantity sold for one day as a group

plt.figure(figsize=(10, 6))

df_group.plot(kind='line', marker='o')

plt.title('Sales Trend Over Time')

plt.xlabel('Date')

plt.ylabel('Total Quantity Sold')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()

```

```python
df_grouped = df.groupby('Date')['UnitPrice'].sum()   # trying to see the trend of low price & high price items which will decide the overall profitability for the business before actually calculating the total sales amount

plt.figure(figsize=(10, 6))

df_grouped.plot(kind='line', marker='o')

plt.title('Sales Trend Over Time')

plt.xlabel('Date')

plt.ylabel('UnitPriceSum')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```
